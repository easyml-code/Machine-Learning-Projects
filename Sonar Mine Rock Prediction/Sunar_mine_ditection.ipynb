```python
import numpy as np
import pandas as pd

data = pd.read_csv('sonar_all_data.csv', header = None)
data.head()
```

```python
X_data = np.array(data)[:,:60]
y_data = np.array(data)[:,-1]
X_data, y_data, X_data.shape, y_data.shape
```

```python
data.describe()
```

```python
data[60].value_counts() #This will show the numbers of data available for each classes
```

```python
# M --> Mine
# R --> Rock

X = data.drop( columns = 60, axis = 1)
y = data[60]
# y
# y_data = len(y)*[0]
y_data = []
for i in range(len(y)):
    if y[i] == 'R':
        temp = 0
    else:
        temp = 1
    y_data.append(temp)
        
X, y_data
```

```python
from sklearn.model_selection import train_test_split
np.random.seed(1)
X_train, X_test, y_train, y_test = train_test_split(X, y_data, test_size = 0.1, stratify = y, random_state = 1)
X_train.shape, X_test.shape
# , y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_percentage_error as MAPE
from sklearn import preprocessing

model_mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(10,8), activation = 'tanh', solver = 'lbfgs', random_state=10)
model_mlp.fit(X_train, y_train)
```

```python
from sklearn import svm
model_svm = svm.SVC()
```

```python
# y_data=y_data.astype('int')
# y_train=y_train.astype('int')

model.fit(X_train, y_train)
model_mlp.fit(X_train, y_train)
model_svm.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
# y_pred, y_test
```

```python
y_pred2 = model_mlp.predict(X_test)
```

```python
y_pred3 = model_svm.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# y_test = y_test.astype('int')
print(accuracy_score(y_test, y_pred))
print(accuracy_score(y_test, y_pred2))
print(accuracy_score(y_test, y_pred3))
print(accuracy_score(y_train, model.predict(X_train)))
```
